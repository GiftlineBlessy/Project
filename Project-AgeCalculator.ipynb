```python
import gradio as gr
import datetime as dt

def calculate_age(birthdate):
    try :
        birth_date = dt.datetime.strptime(birthdate, "%Y-%m-%d")
        today = dt.datetime.today()
        if((today.month, today.day) < (birth_date.month, birth_date.day)):
                  age = today.year - birth_date.year - 1
        else:
         age =today.year-birth_date.year
        return f"You are {age} years old."
    except ValueError:
        return "Invalid format. Please enter your birth date in YYYY-MM-DD format."

print ("Age Calculator")

while True:
    birth_input = input("Enter your date of birth (YYYY-MM-DD): ")
    Age = calculate_age(birth_input)
    print (Age)
    break
```

```python
import gradio as gr
gr.Interface(
    fn = calculate_age,
    inputs = gr.Textbox(label = "Enter your birth date (YYYY-MM-DD): "),
    outputs = "text",
    title = "Age Calculator",
).launch()
```
